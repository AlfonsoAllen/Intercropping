```python
import matplotlib
matplotlib.use('TkAgg')
from pylab import *
import numpy as np
import csv
from scipy import interpolate
```

```python

def extraer_valores(lista_beta,lista_R):
    
    f = interpolate.interp1d(lista_beta,lista_R)
    lista_beta_adap = np.linspace(lista_beta[0], lista_beta[-1], num=1000)
    lista_R_adap = f(lista_beta_adap)
    
    k = 0

    while lista_R_adap[k] < 0.0025 and k < len(lista_R_adap)-1:

        k += 1
    a10 = lista_beta_adap[k]  
    
    return a10

def extraer_valores2(lista_beta,lista_R):
    
    f = interpolate.interp1d(lista_beta,lista_R)
    lista_beta_adap = np.linspace(lista_beta[0], lista_beta[-1], num=1000)
    lista_R_adap = f(lista_beta_adap)
    
    k = 0

    while lista_R_adap[k] < 0.0012 and k < len(lista_R_adap)-1:

        k += 1
    a10 = lista_beta_adap[k]  
    
    return a10
```
