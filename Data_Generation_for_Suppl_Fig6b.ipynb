```python
from pylab import *
import random as rd
import numpy as np
import networkx as nx
```

```python
def rectangular_geometric_lattice(F,C,r):
    
    a = ((C - 1.0) / (F - 1.0))**0.5
    d = 1.0/(a * (F-1))

    g = nx.Graph()
    
    pos_x = 0.0
    pos_y = 0.0
    
    for i in range(0,int(F*C)):
        
        if i == 0:

            g.add_node(i)
            g.node[i]['x'] = pos_x
            g.node[i]['y'] = pos_y
            
        else:
            
            g.add_node(i)
            
            if i % C == 0.0:
                
                pos_y += d
                pos_x = 0.0
                
                g.node[i]['x'] = pos_x
                g.node[i]['y'] = pos_y
                
            else:
                
                pos_x += d
                
                g.node[i]['x'] = pos_x
                g.node[i]['y'] = pos_y
            
           

    for i in g.nodes_iter():

        for j in g.nodes_iter():

            if i != j and g.has_edge(i,j)==False and\
            (g.node[i]['x'] - g.node[j]['x'])**2.0 + (g.node[i]['y']-g.node[j]['y'])**2.0 <= r**2.0:

                g.add_edge(i,j)
                
    return g

```

```python
def monocrop(F,C,fraction_infected_t0,inf_border):
    
    crop_inmune = []
    crop_suscep = []
    
    for i in range(0,int(F*C)):
            
        crop_suscep.append(i)
            
    if inf_border == False:
        
        crop_infect = rd.sample(crop_suscep,int(fraction_infected_t0*len(crop_suscep)))
        
    else:
        
        borde_rec = []
        
        for i in range(0,int(C)):
            
            borde_rec.append(i)
        
        for i in range(int((F-1)*C),int(C*F)):
            
            borde_rec.append(i)
        
        for i in range(1,int(F-1)):
            
            borde_rec.append(int(C*i))
            borde_rec.append(int(C*i)+int(C-1))
            
        borde_sus = [i for i in borde_rec if i not in crop_inmune]
        crop_infect = rd.sample(borde_sus,int(1))
    
    return crop_inmune, crop_suscep, crop_infect

def strips(F,C,fraction_infected_t0,inf_border): #Pensado para F=20, C= 50
    
    plantas = range(0,int(F*C))
    
    crop_inmune = []
    
    filas = [0,3,4,7,8,11,12,15,16,19]

    for i in filas:
            
        for j in range(int(i*C),int(((i+1)*C))):

            crop_inmune.append(j)
        
    crop_suscep = [x for x in plantas if x not in crop_inmune]
    
    if inf_border == False:
        
        crop_infect = rd.sample(crop_suscep,int(fraction_infected_t0*len(crop_suscep)))
        
    else:
        
        borde_rec = []
        
        for i in range(0,int(C)):
            
            borde_rec.append(i)
        
        for i in range(int((F-1)*C),int(C*F)):
            
            borde_rec.append(i)
        
        for i in range(1,int(F-1)):
            
            borde_rec.append(int(C*i))
            borde_rec.append(int(C*i)+int(C-1))
            
        borde_sus = [i for i in borde_rec if i not in crop_inmune]
        crop_infect = rd.sample(borde_sus,int(1))
    
    return crop_inmune, crop_suscep, crop_infect

def random(F,C,fraction_infected_t0,inf_border):
    
    plantas = range(0,int(F*C))
    
    crop_inmune = rd.sample(plantas,int(0.5*F*C))
    crop_suscep = [i for i in plantas if i not in crop_inmune]
    
    if inf_border == False:
        
        crop_infect = rd.sample(crop_suscep,int(fraction_infected_t0*len(crop_suscep)))
        
    else:
        
        borde_rec = []
        
        for i in range(0,int(C)):
            
            borde_rec.append(i)
        
        for i in range(int((F-1)*C),int(C*F)):
            
            borde_rec.append(i)
        
        for i in range(1,int(F-1)):
            
            borde_rec.append(int(C*i))
            borde_rec.append(int(C*i)+int(C-1))
            
        borde_sus = [i for i in borde_rec if i not in crop_inmune]
        crop_infect = rd.sample(borde_sus,int(1))
    
    return crop_inmune, crop_suscep, crop_infect

def row(F,C,fraction_infected_t0,inf_border):
    
    a = ((C - 1.0) / (F - 1.0))**0.5
    d = 1.0/(a * (F-1))
    
    crop_inmune = []
    crop_suscep = []
    
    for i in range(0,int(F)):
            
        if 1.0*int(i) % 2 == 0:
            
            crop_suscep += range(i*C,(i+1)*int(C))
            #print i,1.0*int(C) % 2, 'sus',crop_suscep
            
        else:
            
            crop_inmune += range(i*C, (i+1)*int(C))
            #print i,1.0*int(C) % 2, 'inm',crop_inmune
    
    if inf_border == False:
        
        crop_infect = rd.sample(crop_suscep,int(fraction_infected_t0*len(crop_suscep)))
        
    else:
        
        borde_rec = []
        
        for i in range(0,int(C)):
            
            borde_rec.append(i)
        
        for i in range(int((F-1)*C),int(C*F)):
            
            borde_rec.append(i)
        
        for i in range(1,int(F-1)):
            
            borde_rec.append(int(C*i))
            borde_rec.append(int(C*i)+int(C-1))
            
        borde_sus = [i for i in borde_rec if i not in crop_inmune]
        crop_infect = rd.sample(borde_sus,int(1))
    
    return crop_inmune, crop_suscep, crop_infect

def columna(F,C,fraction_infected_t0,inf_border):
    
   
    crop_inmune = []
    crop_suscep = []
    
    for i in range(0,int(C*F)):
            
        if 1.0*int(i) % 2 == 0:
            
            crop_suscep.append(i)
            #print i,1.0*int(C) % 2, 'sus',crop_suscep
            
        else:
            
            crop_inmune.append(i)
            #print i,1.0*int(C) % 2, 'inm',crop_inmune
    
    if inf_border == False:
        
        crop_infect = rd.sample(crop_suscep,int(fraction_infected_t0*len(crop_suscep)))
        
    else:
        
        borde_rec = []
        
        for i in range(0,int(C)):
            
            borde_rec.append(i)
        
        for i in range(int((F-1)*C),int(C*F)):
            
            borde_rec.append(i)
        
        for i in range(1,int(F-1)):
            
            borde_rec.append(int(C*i))
            borde_rec.append(int(C*i)+int(C-1))
            
        borde_sus = [i for i in borde_rec if i not in crop_inmune]
        crop_infect = rd.sample(borde_sus,int(1))
    
    return crop_inmune, crop_suscep, crop_infect

def chessboard(F,C,fraction_infected_t0,inf_border):
   
    crop_inmune = []
    crop_suscep = []
    
    for i in range(0,int(F)):
            
        for j in range(0,int(C)):
            
            if i % 2 == 0:
                
                if j % 2 == 0:
            
                    crop_suscep.append(i*C+j)
                
                else:
                    
                    crop_inmune.append(i*C+j)
                
            else:
                
                if j % 2 == 0:
            
                    crop_inmune.append(i*C+j)
                
                else:
                    
                    crop_suscep.append(i*C+j)
    
    if inf_border == False:
        
        crop_infect = rd.sample(crop_suscep,int(fraction_infected_t0*len(crop_suscep)))
        
    else:
        
        borde_rec = []
        
        for i in range(0,int(C)):
            
            borde_rec.append(i)
        
        for i in range(int((F-1)*C),int(C*F)):
            
            borde_rec.append(i)
        
        for i in range(1,int(F-1)):
            
            borde_rec.append(int(C*i))
            borde_rec.append(int(C*i)+int(C-1))
            
        borde_sus = [i for i in borde_rec if i not in crop_inmune]
        crop_infect = rd.sample(borde_sus,int(1))
    
    return crop_inmune, crop_suscep, crop_infect

def patches(F,C,fraction_infected_t0,Fp,Cp,inf_border):
   
    crop_inmune = []
    crop_suscep = []
    
    lista_Fp = [i for i in range(0,int(F)) if i % Fp == 0]
    lista_Cp = [i for i in range(0,int(C)) if i % Cp == 0]
    
    lista_Fp_tipo = []
    lista_Cp_tipo = []
    
    for i in range(0,len(lista_Fp)):
        
        if i%2==0:
            
            lista_Fp_tipo.append(1.0) #1 es susceptible
            
        else:
            
            lista_Fp_tipo.append(0.0) #0 es inmune
    
    for i in range(0,len(lista_Cp)):
        
        if i%2 == 0:
            
            lista_Cp_tipo.append(1.0) #1 es susceptible
            
        else:
            
            lista_Cp_tipo.append(0.0) #0 es inmune

    for i in range(0,len(lista_Cp)):
        
        tipo_x = lista_Cp_tipo[i]
        
        for x in range(lista_Cp[i],lista_Cp[i]+Cp):
            
            for j in range(0,len(lista_Fp)):
                
                tipo_y = lista_Fp_tipo[j]
                
                for y in range(lista_Fp[j],lista_Fp[j]+Fp):
                    
                    if tipo_y==tipo_x:
                        
                        crop_suscep.append(y*C+x)
                        #print x,y,y*F+x,'sus'
                        
                    else:
                        
                        crop_inmune.append(y*C+x)
                        #print x,y,y*F+x,'inm'
   
    if inf_border == False:
        
        crop_infect = rd.sample(crop_suscep,int(fraction_infected_t0*len(crop_suscep)))
        
    else:
        
        borde_rec = []
        
        for i in range(0,int(C)):
            
            borde_rec.append(i)
        
        for i in range(int((F-1)*C),int(C*F)):
            
            borde_rec.append(i)
        
        for i in range(1,int(F-1)):
            
            borde_rec.append(int(C*i))
            borde_rec.append(int(C*i)+int(C-1))
            
        borde_sus = [i for i in borde_rec if i not in crop_inmune]
        crop_infect = rd.sample(borde_sus,int(1))
    
    return crop_inmune, crop_suscep, crop_infect
```

```python
def q_i_SIN(nodo,lista_p_i):
    
    global g,beta_cons

    aux = 1.0
    
    for i in g.neighbors_iter(nodo):
        
        aux *= (1.0 - beta_cons*lista_p_i[i])
        
    return 1.0 - aux

def Markov_evolution_SIR_SIN_cuad(crop_inmune, crop_suscep, crop_infect):
    
    global g,beta_cons,mu_cons,tmax,lista_infectados_ini,\
    lista_recuperados_ini,lista_recuperados_final
    
    N0 = len(g.nodes())
    
    lista_s_i = [] #susceptibles
    lista_p_i = [] #infectados
    lista_rho_i = [] #recuperados
    
    for i in g.nodes_iter():

        lista_s_i.append(1.0) #susceptibles
        lista_p_i.append(0.0) #infectados
        lista_rho_i.append(0.0) #recuperados
        
    N0 = len(g.nodes())
    
    cant_inmunes_t0 = len(crop_inmune)
    lista_infectados_ini = crop_infect[:]
    lista_recuperados_ini = crop_inmune[:]
    
    #print 'inmunes:',1.0*len(crop_inmune)/N0,'infectados:',1.0*len(crop_infect)/len(crop_suscep)

    for i in crop_inmune:
        
        lista_s_i[i] =  0.0
        lista_p_i[i] = 0.0
        lista_rho_i[i] = 1.0
        
    for i in crop_infect :
        
        lista_s_i[i] =  0.0
        lista_p_i[i] = 1.0
        lista_rho_i[i] = 0.0
        
    
    amount_inf.append(1.0*sum(lista_p_i)/(N0-cant_inmunes_t0))
    amount_sus.append(1.0*sum(lista_s_i)/(N0-cant_inmunes_t0))
    amount_rec.append(1.0*(sum(lista_rho_i)-cant_inmunes_t0)/(N0-cant_inmunes_t0))
    
    t = 0
    terminar = False
    
    while t <= tmax and terminar==False:
    
        new_s_i = []
        new_p_i = []
        new_rho_i = []

        for a in g.nodes_iter():

        
            if lista_rho_i[a] == 1.0:
                
                new_s_i.append(0.0)
                new_p_i.append(0.0)
                new_rho_i.append(1.0)
                
            else:  

                q_i_nodo = q_i_SIN(a,lista_p_i)

                aux_new_p_i = lista_p_i[a]*(1.0 - mu_cons) + (1.0 - lista_p_i[a] - lista_rho_i[a])*q_i_nodo
                aux_new_rho_i = lista_rho_i[a] + mu_cons*lista_p_i[a]

                new_s_i.append(1.0 - aux_new_p_i - aux_new_rho_i)
                new_p_i.append(aux_new_p_i)
                new_rho_i.append(aux_new_rho_i)
        
        if max(new_p_i) < 1e-6:

            terminar = True

        lista_s_i = new_s_i[:]
        lista_p_i = new_p_i[:]
        lista_rho_i = new_rho_i[:]
        
        
        amount_inf.append(1.0*sum(lista_p_i)/(N0-cant_inmunes_t0))
        amount_sus.append(1.0*sum(lista_s_i)/(N0-cant_inmunes_t0))
        amount_rec.append(1.0*(sum(lista_rho_i)-cant_inmunes_t0)/(N0-cant_inmunes_t0))

        t += 1
        
    
    #print 't_sim:',t
    
    lista_recuperados_final=[i for i in g.nodes_iter() if lista_rho_i[i] > lista_s_i[i]]
    
    return 1.0*(sum(lista_rho_i)-cant_inmunes_t0)/(N0-cant_inmunes_t0)
```

```python
def matriz_distancia():
    
    global g
    
    dic_shortest = nx.shortest_path(g)

    M_nueva = []

    for i in range(0,len(g.nodes())):

        fila_aux = []

        fila = dic_shortest.get(i)

        for k in range(0,len(g.nodes())):

            fila_aux.append(len(fila.get(k))-1.0)

        M_nueva.append(fila_aux)
    
    return M_nueva
```

```python
def matriz_alargamiento():
    
    global g,crop_inmune
    
    dic_shortest = nx.shortest_path(g)

    M_nueva = []

    for i in range(0,len(g.nodes())):

        fila_aux = []

        fila = dic_shortest.get(i)

        for k in range(0,len(g.nodes())):
            
            camino_ik = fila.get(k)
            
            aux = 0.0
            
            for p in camino_ik:
                
                if p in crop_inmune:
                    
                    aux += 1.0

            fila_aux.append(aux)

        M_nueva.append(fila_aux)
    
    return M_nueva
```

```python
def q_i_MELLIN_SEXY3(nodo,lista_p_i,s):
    
    global g,beta_cons,M_distancia,M_alargamiento,crop_suscep

    aux = 1.0
    
    for i in crop_suscep:

        if i != nodo:
            
            if s >= 100 and M_alargamiento[nodo][i] == 0.0:

                aux *= (1.0 - 1.0*beta_cons*lista_p_i[i])
                
            elif s < 100 and M_alargamiento[nodo][i] == 0.0:

                aux *= (1.0 - 1.0*beta_cons*lista_p_i[i]/((M_distancia[nodo][i])**s))
                
    return (1.0 - aux)
    

def Markov_evolution_SIR_MELLIN_SEXY_cuad3_time(s,crop_inmune, crop_suscep, crop_infect):
    
    global g,beta_cons,mu_cons,tmax,lista_infectados_ini,\
    lista_recuperados_ini,lista_recuperados_final,M_distancia,M_alargamiento,rho_est
    
    N0 = len(g.nodes())
    
    lista_s_i = [] #susceptibles
    lista_p_i = [] #infectados
    lista_rho_i = [] #recuperados
    
    for i in g.nodes_iter():

        lista_s_i.append(1.0) #susceptibles
        lista_p_i.append(0.0) #infectados
        lista_rho_i.append(0.0) #recuperados
        
    N0 = len(g.nodes())
    
    cant_inmunes_t0 = len(crop_inmune)
    lista_infectados_ini = crop_infect[:]
    lista_recuperados_ini = crop_inmune[:]
    
    #print 'inmunes:',1.0*len(crop_inmune)/N0,'infectados:',1.0*len(crop_infect)/len(crop_suscep)


    for i in crop_inmune:
        
        lista_s_i[i] =  0.0
        lista_p_i[i] = 0.0
        lista_rho_i[i] = 1.0
        
    for i in crop_infect :
        
        lista_s_i[i] =  0.0
        lista_p_i[i] = 1.0
        lista_rho_i[i] = 0.0

    
    amount_inf.append(1.0*sum(lista_p_i)/(N0-cant_inmunes_t0))
    amount_sus.append(1.0*sum(lista_s_i)/(N0-cant_inmunes_t0))
    amount_rec.append(1.0*(sum(lista_rho_i)-cant_inmunes_t0)/(N0-cant_inmunes_t0))
    
    
    t = 0
    terminar = False
    
    while t <= tmax and terminar==False:
    
        new_s_i = []
        new_p_i = []
        new_rho_i = []

        for a in g.nodes_iter():
            
            if g.node[a]['buscar'] == True: #lista_p_i[a] < lista_rho_i[a]:
                
                g.node[a]['time'] = t
                
                if 1.0 - lista_p_i[a] - lista_rho_i[a] < lista_rho_i[a]:
                
                    g.node[a]['buscar'] = False
            
            if s >= 100:
            
                q_i_nodo = q_i_SIN(a,lista_p_i)
                
            else:
                
                q_i_nodo = q_i_MELLIN_SEXY3(a,lista_p_i,s) #q_i_MELLIN(a,lista_p_i,s)

            aux_new_p_i = lista_p_i[a]*(1.0 - mu_cons) + (1.0 - lista_p_i[a] - lista_rho_i[a])*q_i_nodo
            aux_new_rho_i = lista_rho_i[a] + mu_cons*lista_p_i[a]

            new_s_i.append(1.0 - aux_new_p_i - aux_new_rho_i)
            new_p_i.append(aux_new_p_i)
            new_rho_i.append(aux_new_rho_i)

        
        if max(new_p_i) < 1e-6:

            terminar = True

        lista_s_i = new_s_i[:]
        lista_p_i = new_p_i[:]
        lista_rho_i = new_rho_i[:]
        
        amount_inf.append(1.0*sum(lista_p_i)/(N0-cant_inmunes_t0))
        amount_sus.append(1.0*sum(lista_s_i)/(N0-cant_inmunes_t0))
        amount_rec.append(1.0*(sum(lista_rho_i)-cant_inmunes_t0)/(N0-cant_inmunes_t0))

        t += 1
        
    
    
    return 1.0*(sum(lista_rho_i)-cant_inmunes_t0)/(N0-cant_inmunes_t0)
```
